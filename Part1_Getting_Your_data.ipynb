```python
df_csv_title = pd.read_csv("IMDB-Movie-Data.csv" , index_col= "Title")
```

### Removed char and convert lower case

```python
df_csv_title.columns = ['rank' , 'genre' ,'desc' ,'dir' ,'actor' ,'year' ,'length' ,'rating' ,'votes' ,'rev' ,'mscore']
```

```python
df_csv_title.index.name = 'title'
```

```python
df_csv_title.head(2)
```

```python
df_csv_title.info()
```

```python
df_csv_title.info
```

```python
df_csv_title.shape
```

```python
df_csv_title.describe()
```

### Find null and replace with mean value

```python
df_csv_title.isnull().sum()
```

```python
df_csv_title['rev'].fillna(df_csv_title['rev'].mean() , inplace = True)
```

```python
df_csv_title.isnull().sum()
```

```python
df_csv_title['mscore'].fillna(df_csv_title['mscore'].mean() , inplace = True)
```

```python
df_csv_title.isnull().sum()
```

### inplace option will make chnage to current data frame 

```python
df_csv_title.drop(['rank'], axis=1 , inplace=True)
```

```python
df_csv_title.head(2)
```

```python
df_csv_title.columns
```

```python
df_csv_title.groupby(['year']).mean()
```

```python
df_csv_title.groupby(['year']).mean().sort_values(by=['rev'], ascending = False)
```

```python
df_csv_title[df_csv_title['year'] == 2012].sort_values(by=['rev'], ascending=False)
```

```python
df_csv_title[df_csv_title['year'] == 2012].sort_values(by=['rev'], ascending=False)
```

```python
df_csv_title[df_csv_title['genre'].isin(['Comedy','Action'])].sort_values(by=['rating'],ascending= False)
```

# Create Profile Report 

```python
!pip install pandas-profiling[notebook,html]
```

```python
import numpy as np
import pandas as pd
from pandas_profiling import ProfileReport
```

```python
pd_mv_csv = pd.read_csv('IMDB-Movie-Data.csv')
```

```python
profile = ProfileReport(pd_mv_csv, title='Pandas Profiling Report', html={'style':{'full_width':True}})
```

```python
profile
```

```python
profile.to_notebook_iframe()
```

## Save Report

```python
profile.to_file(output_file="your_report.html")
```

### Alternatively, you can obtain the data as json:

```python
json_data = profile.to_json()

# As a file
profile.to_file(output_file="your_report.json")
```

### Large datasets 
Version 2.4 introduces minimal mode. This is a default configuration that disables expensive computations (such as correlations and dynamic binning). 

```python
pd_mv_csv_lard_data = pd.read_csv('IMDB-Movie-Data.csv')

profile = ProfileReport(pd_mv_csv_lard_data, minimal=True)
profile.to_file(output_file="output.html")
```
